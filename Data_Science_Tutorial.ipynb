```python
import matplotlib.pyplot as plt
year=[1,2,3]
veg_rate=[2,4,6]
milk_rate=[1,3,5]
fruit_rate=[3,6,9]
line1,=plt.plot(year,veg_rate)
line2,=plt.plot(year,milk_rate)
line3,=plt.plot(year,fruit_rate)
plt.legend([line1,line2,line3],['vegetable','milk','fruit'])
plt.xlabel('years')
plt.ylabel('rate')
plt.show()
print('successfully placed a legend on the axes....')
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['lion','horse','kangaroo']
values1=[15,24,30]
values2=[20,18,25]
bar_width=0.35
bar_positions1=np.arange(len(categories))
bar_positions2=bar_positions1+bar_width
plt.bar(bar_positions1,values1,width=bar_width,label='2021',color='skyblue')
plt.bar(bar_positions2,values2,width=bar_width,label='2022',color='blue')
plt.xlabel('categories')
plt.ylabel('values')
```

```python
import matplotlib.pyplot as plt
categories=['transport','income tax','housing']
values1=[25,24,30]
values2=[20,18,25]
plt.bar(categories,values1,label='first half',color='skyblue')
plt.bar(categories,values2,label='second half',color='pink')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('stackedd bar graph')
plt.legend()
plt.show()
```

```python
#cummulative histogram
import matplotlib.pyplot as plt
import numpy as np
exam_scores=np.random.randint(0,100,100)
plt.hist(exam_scores,bins=10,cumulative=True,color='blue',edgecolor='black')
plt.xlabel('exam_scores')
plt.ylabel('no of students')
plt.title('cumulative histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
mark1=np.array([2,4,5,6,7,8,9,10,13,15])
mark2=np.array([3,4,5,6,8,9,10,13,15,16])
plt.hist(mark1,bins=5,stacked=True,color='skyblue',edgecolor='black')
plt.hist(mark2,bins=5,stacked=True,color='pink',edgecolor='brown')
plt.title('stacked histogram')
plt.xlabel('marks')
plt.ylabel('no of students')
plt.legend(['mark dataset1','mark dataset2'])
plt.show()
```

```python
#pandas library
#it has two datatypes they are series and data frame
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(a)
print(myvar)
print(myvar[0])
print('.....................')
myvar2=pd.Series(a,index=['x','y','z'])
print(myvar2)
print(myvar2['y'])
print('.......................')
#generating for a dictonary
calories={'day1':420,'day2':280,'day3':390}
myvar=pd.Series(calories)
print(myvar)
print('.................')
myvar=pd.Series(calories,index=['day1','day2'])
print(myvar)
print('................')
#Series generated from scalar value
series=pd.Series(5,index=[0,1,2,3])
print(series)
print('................')
```

```python
#interger series
int_series=pd.Series([1,2,3,4])
print(int_series)
print('.................')
#float series
float_series=pd.Series([1.2,2.2,3.3,4.4])
print(float_series)
print('.....................')
#string series
str_series=pd.Series(['a','b','c','d'])
print(str_series)
print('.....................')
#label based indexing
data={'a':1,'b':2,'c':3}
series=pd.Series(data)
print(series)
print(series['b'])
print(series[['a','c']])
print('.........................')
#boolean indexing
data=[10,20,30,40,50]
series=pd.Series(data)
#access elements greater than 25
print(series[series>25])
```

```python
#data frame
data={
    'name':['alice','bob','charlie'],
    'age':[25,30,35],
    'city': ['new york','los angeles','chicago']
}
df=pd.DataFrame(data)
print(df)
#selecting columns
print(df.name)
print(df['name'])
print(df[['name','age']])
#select a single row by label
print(df.loc[0])
print('.....................')
print(df.iloc[0])
print('.................')
#select multiple rows
print(df.iloc[0:2])
#coditional selection
print(df[df['age']>25])
print(".......................")
print(df.loc[(df.name=="bob")&(df.city=="los angeles")])

```

```python
data={
    'name':['alice','bob','charlie','alice','bob'],
    'age':[25,30,35,25,30],
    'city': ['new york','san francisco','los angeles','chicago','san francisco'],
    'salary':[70000,800000,1200000,1500000,170000]
}
df=pd.DataFrame(data)
grouped=df.groupby('name')
print(grouped.groups)
#calculate the sum of salaries for each group
salary_sum=grouped['salary'].sum()
print(salary_sum)
age_mean=grouped['age'].mean()
print(age_mean)
#grouping by multiple columns
grouped_multiple=df.groupby(['name','city'])
salary_sum_multiple=grouped_multiple['salary'].sum()
print(salary_sum_multiple)
```
