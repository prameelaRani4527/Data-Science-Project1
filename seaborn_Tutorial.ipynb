```python
#seaborn is a advanced library for plotting. it consists of in-built datasets
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
print(df.to_string())
g=sb.FacetGrid(df,col='time')
g.map(plt.hist,'tip')
plt.show()
```

```python
df=sb.load_dataset('tips')
g=sb.FacetGrid(df,col='sex',hue="smoker")
g.map(plt.scatter,'total_bill','tip')
plt.show()
g=sb.FacetGrid(df,col='time',hue="sex")
g.map(plt.scatter,'total_bill','tip')
```

```python
#statistical estimation
sb.barplot(x='sex',y='tip',hue='time',data=df)
plt.show()
```

```python
#linear relationship
sb.regplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
#pair plot
import seaborn
seaborn.pairplot(df)
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
df=sb.load_dataset('tips')
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
#adding hue color to a seaborn pairplot
sb.pairplot(df,hue='size')
plt.show()
```

```python
sb.pairplot(df,hue='size',palette='husl')
plt.show()
```

```python
#diagonla kind of plots
sb.pairplot(df,diag_kind='kde')#keneral distribution estimator
plt.show()
```

```python
#adjusting plot kind
sb.pairplot(df,kind='reg')
plt.show()
```
